For example, a model that forecasts low-priced cars and one that forecasts high-priced cars can be compared on the same scale, making it easier to identify which model performs better regardless of the price level. Furthermore, MAPE encourages robust model development because it is sensitive to outliers.
```python
# Create a subset of categorical columns from the training data
cat_train = train[['transmission_display', 'wheel_system', 'city', 'body_type',
                   'exterior_color', 'engine_type', 'franchise_dealer', 'is_new', 'listed_date',
                   'make_name', 'model_name', 'listing_color', 'transmission', 'interior_color',
                   'fuel_type']].astype(str)



for col in cat_train.columns:
    # If the column has more than 5 unique values:
    if cat_train[col].nunique() > 5:
        value_counts_train = cat_train[col].value_counts()
        # Calculate value counts and get the top 5 most frequent values
        top_5_val_train = value_counts_train.head(5).index.tolist()
        # Replace values not in the top 5 with 'Other'
        cat_train[col] = cat_train[col].apply(lambda x: x if x in top_5_val_train else 'Other')

# Update the categorical columns in the original training DataFrame
train[cat_train.columns] = cat_train
```

### **Machine Learning Models**

**Gradient Boost Regressor**

The Gradient Boost Regressor Model is a machine learning ensemble technique that combines the predictions of weak learners, specifically decision trees. The aim is to improve the overall predictive performance by optimising the model's weights based on previous iterations. Each tree attempts to correct the errors of previous trees. Decision tree models are often well-suited for handling non-linear relationships.

**Lasso Regression**

Lasso Regression is an L1 regularisation technique that applies a penalty known as alpha ($\lambda$). This penalty is applied to prevent overfitting and enhance the model's accuracy. This model uses something called "shrinkage". Shrinkage is where data values are shrunk towards a central point.

**Polynomial Regression**

Polynomial Regression is a form of regression analysis in which a relationship exists between a dependent variable  and one or more independent variables . This form of Linear Regression models the relationship between the variables as an n-degree polynomial, allowing non-linear relationships to be captured.


```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_percentage_error
gradient_regressor = GradientBoostingRegressor(
    n_estimators=10,
    learning_rate=0.1,
    max_depth=10,
    min_samples_leaf=3,
    random_state=42
)

X = train.drop('price', axis=1)
y = train['price']

X_test = test[test.columns]


gradient_regressor.fit(X, y)


training_predictions = gradient_regressor.predict(X)

mape = mean_absolute_percentage_error(y, training_predictions) * 100
print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
```

### **Lasso Regression**

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=1.0, random_state=1)

lasso_reg.fit(X, y)

train_pred_lasso = lasso_reg.predict(X)

mape = mean_absolute_percentage_error(y, train_pred_lasso) * 100
print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
```

### **Polynomial Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X_poly = PolynomialFeatures(degree=2).fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

train_pred_poly = poly_reg.predict(X_poly)

mape = mean_absolute_percentage_error(y, train_pred_poly) * 100
print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
```
